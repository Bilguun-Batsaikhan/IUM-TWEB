### Definition for all columns ###

+ club_id: A unique identifier for each club.
+ club_code: A code or abbreviation representing each club.
+ name: The complete name of the club.
+ domestic_competition_id: A unique identifier for each competition.
+ total_market_value: This value is null for every column.
+ squad_size: It is the number of players for every club.
+ average_age: It is the mean age of players for every club.
+ foreigners_number: It is the number of foreign players for every club.
+ foreigners_percentage: It is the percentage of foreign players for every club.
+ national_team_players: It is the number of players for every club that also play for a national team.
+ stadium_name: This is the name of the stadium for every club.
+ stadium_seats: This is the number of seats in the stadium for every club.
+ net_transfer_record: It is the balance of the transfer market.
+ coach_name: This value is null for every column.
+ last_season: The year of the last games played by that club in a competitions.
+ url: The URL link to access more information about the club, potentially leading to details such as schedules, results, and standings.

### Step 2: Data Preperation ###
+ Dropping irrelevant columns and rows (NaN)
+ Identifying duplicated columns
+ Feature Creation

```python
club.isna().sum() #total_market_value && coach_name are null for every row
```

```python
club.loc[club['average_age'].isna()] #these row can be ignored because there isn't significance data 
```

```python
club = club.loc[~club['average_age'].isna()] #remove these row
```

```python
club.isna().sum()
```

```python
club.loc[club['foreigners_percentage'].isna()] #these row have 0 foreigners_number, so simply change NaN value with 0
```

```python
club['foreigners_percentage'].fillna(0, inplace=True)
club['foreigners_percentage'].isna().sum()
```

```python
club['net_transfer_record'].describe()
```

```python
club['net_transfer_record'].unique()
```

```python
club['net_transfer_record'] = club['net_transfer_record'].str.replace('€', '').str.replace('m', 'e6').str.replace('k', 'e3').str.replace('+-', '').astype(float) # remove symbol €, 'm','k' and '+-', trasform from string to float value
club['net_transfer_record'].describe()
```
